```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # colourmap module
import warnings

%matplotlib inline 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("C:/Users/Karan Jadhav/OneDrive/Desktop/Prodigy_Infotech_DS/Prodigy_Infotech_DS_Task_01/PopulationDataset.csv", skiprows=4)
column_idx = df.columns.get_loc("2023")
df = df.iloc[:, :column_idx+1]
```

```python
countries_to_exclude = [
    "Low & middle income","Middle income","IBRD only","Early-demographic dividend","Upper middle income", "High income", "Low income",
    "Lower middle income", "Late-demographic dividend", "Post-demographic dividend", "IDA & IBRD total", "IDA total", "IDA blend",
    "IDA only", "Euro area", "East Asia & Pacific (excluding high income)","East Asia & Pacific","Europe & Central Asia (excluding high income)",
    "Europe & Central Asia","Latin America & Caribbean (excluding high income)","Latin America & Caribbean","Middle East & North Africa",
    "Middle East & North Africa (excluding high income)","North America","South Asia","Sub-Saharan Africa (excluding high income)","Sub-Saharan Africa",
    "World","Arab World", "Central Europe and the Baltics","Caribbean small states","Central African Republic", "Channel Islands",
    "Developing small island states","East Asia & Pacific (IDA & IBRD countries)","Europe & Central Asia (IDA & IBRD countries)",
    "Fragile and conflict affected situations","Heavily indebted poor countries (HIPC)","High income: nonOECD","High income: OECD",
    "Latin America & the Caribbean (IDA & IBRD countries)","Least developed countries: UN classification","Middle East & North Africa (IDA & IBRD countries)",
    "Pacific island small states","Small states","South Asia (IDA & IBRD)","Sub-Saharan Africa (IDA & IBRD countries)",
    "Other small states","Pre-demographic dividend","European Union","Least Developed Countries","Less Developed Regions",
    "More Developed Regions","Land Locked Developing Countries","Small Island Developing States","Eastern and South-Eastern Asia",
    "Northern Africa and Western Asia","Oceania","Northern America and Europe","Eastern and Southern Africa","Middle Africa",
    "Western Africa","Western Asia","OECD members","Not classified"
]

# Apply the filter to the DataFrame
df = df[~df["Country Name"].isin(countries_to_exclude)]
```

```python
year_columns = [col for col in df.columns if col.isdigit()]
df['Total Population'] = df[year_columns].sum(axis=1, skipna=True)
df
```

```python
# Top 10 Countries by Total Population

top_10_countries = df.sort_values(by = 'Total Population', ascending=False).head(10)
colors = cm.plasma(np.linspace(0, 1, len(top_10_countries)))

plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['Total Population'], color = colors)
plt.xlabel('Name of Countries')
plt.ylabel('Total Population')
plt.title('Top 10 Countries by Total Population')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
# India Population Growth (2015-2023)

india = df[df['Country Name'] == 'India']
plt.figure(figsize=(8, 4))

recent_year_cols = [col for col in year_columns if int(col) >= 2015]

plt.plot(recent_year_cols, india[recent_year_cols].values.flatten(), label='India', color='red')
plt.title("India Population Growth (2015â€“2023)")
plt.xlabel("Year")
plt.ylabel("Population") 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Population comparison of India, USA & China Overtime

countries = df[df['Country Name'].isin(['India', 'China', 'United States'])]
plt.figure(figsize=(10, 5))
for _, row in countries.iterrows():
    plt.plot(year_columns, row[year_columns], label=row['Country Name'])
    
plt.legend()
plt.title('Population comparison of India, USA & China')
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(False)
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
# Population comparison of Three Countries in 2000, 2010, 2020

years_selected = ['2000', '2010', '2020']
data = df[df['Country Name'].isin(['India', 'China', 'United States'])]

data_plot = data.set_index('Country Name')[years_selected].T
data_plot.plot(kind="bar", figsize=(10, 5))
plt.title("Population Comparison (2000, 2010, 2020)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```

```python
# Distribution of population in 2023

plt.figure(figsize=(8, 4))
sns.histplot(df["2023"].dropna(), bins=30, kde=False, color='yellow')
plt.title("Population Distribution in 2023")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
# top 5 countries by population in 2023

top5 = df.sort_values(by='2023', ascending=False).head(5)

top5 = top5.set_index("Country Name")
plt.figure(figsize=(6, 6))
top5["2023"].plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Countries by Population Share (2023)")
plt.tight_layout()
plt.show()
```

```python
# 1980 vs 2023

df_clean = df.dropna(subset=["1980", "2023"])
plt.figure(figsize=(7, 5))
plt.scatter(df_clean["1980"], df_clean["2023"], alpha=0.6)
plt.title("Population: 1980 vs 2023")
plt.xlabel("Population in 1980")
plt.ylabel("Population in 2023")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 5 Countries by population

top_5_countries_list = df.sort_values(by='Total Population', ascending=False).head(5)['Country Name'].tolist()

df_top5 = df[df['Country Name'].isin(top_5_countries_list)]

area_chart_data = df_top5[['Country Name'] + year_columns].set_index('Country Name').T 
area_chart_data = area_chart_data.fillna(0)

plt.figure(figsize=(12, 6))
area_chart_data.plot(kind='area', stacked=True, ax=plt.gca()) 

plt.title("Population Share of Top 5 Countries Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
